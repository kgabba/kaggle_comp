### Выявлена нелинейная связь таргета с фичами: cloud, sunshine

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False)
X_train = pd.concat([X_train, pd.DataFrame(poly.fit_transform(X_train[['sunshine', 'cloud']]), columns=poly.get_feature_names_out())], axis=1).drop(['cloud','sunshine'], axis=1)
```

```python
X_test = pd.concat([X_test, pd.DataFrame(poly.transform(X_test[['sunshine', 'cloud']]), columns=poly.get_feature_names_out())], axis=1).drop(['cloud','sunshine'], axis=1)
```

```python
params = {'n_jobs':-1, 'class_weight':'balanced', 'random_state':42}
logreg2 = LogisticRegression(**params)
```

```python
logreg2.fit(X_train, y_train)
```

```python
roc_auc_score(y_test, logreg2.predict(X_test))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
```

```python
log_pipe = Pipeline([('polyy', PolynomialFeatures(include_bias=False)),
         ('logreg3', LogisticRegression(**params))])
```

```python
rs_params = {'polyy__degree' : [2,3,4,5]}
rs_line_poly = RandomizedSearchCV(log_pipe, rs_params, cv=4,random_state=22, scoring='roc_auc')
```

```python
rs_line_poly.fit(X_train, y_train)
```

```python
rs_line_poly.best_params_
```

```python
rs_line_poly.best_score_
```

```python
svc_pipe = Pipeline([('svc', SVC(random_state=22))])

from sklearn.model_selection import cross_val_score
svc_params = {'svc__degree':[2,3,4,5], 
             'svc__kernel': ['linear', 'poly', 'rbf']}

svc_mod = RandomizedSearchCV(svc_pipe, svc_params, cv=5, n_jobs=-1, scoring='roc_auc')
```

```python
svc_mod.fit(X_train, y_train)
```

```python
svc_mod.best_score_, svc_mod.best_params_
```
